# Label Encoding 
Convert the categorical data into numerical values.

```python
# Import the Libraries
import numpy as np 
import pandas as pd
import sklearn.datasets as datasets
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding on breast Cancer dataset
# loading the breast cancer dataset from sklearn
dataset = datasets.load_breast_cancer()
```

```python
# converting loaded dataset into the pandas DataFrame
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
dataset
```

```python
df.head()
```

```python
df['diagnosis'] = dataset.target
```

```python
df
```

```python
df['diagnosis'] = df['diagnosis'].apply(lambda x: "M" if x == 0 else 'B')
```

```python
df.head()
```

```python
# Finding the count of diagnosis label
df['diagnosis'].value_counts()
```

```python
# Load the label encoder 
le = LabelEncoder()
```

```python
# Fitting the label encoder converting the string labels to numbers (0 & 1)
label = le.fit_transform(df['diagnosis'])
```

```python
# Appending the labels to dataframe
df['target'] = label
```

```python
# 0 -> Benign -> B
# 1 -> Malignant -> M
```

```python
df.head()
```

```python
df.tail()
```

```python
print('Target labels count: ', df['target'].value_counts())
print('Diagnosis labels count: ', df['diagnosis'].value_counts())
```

```python
iris_dataset = datasets.load_iris()
```

```python
iris_dataset
```

```python
iris_df = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
```

```python
iris_df['Species'] = iris_dataset.target
```

```python
iris_df['Species'] = iris_df['Species'].apply(lambda x: 'setosa' if x == 0 else ('versicolor' if x == 1 else 'virginica'))
```

```python
iris_df.head()
```

```python
iris_df.tail()
```

```python
label2 = le.fit_transform(iris_df['Species'])
```

```python
iris_df['Target'] = label2
```

```python
iris_df.head()
```

```python
iris_df.tail()
```

```python
print("The value count: ",iris_df['Species'].value_counts())
print("The value count: ",iris_df['Target'].value_counts())
```
