# MATPLOTLIB 

```python
# Import the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
# Ploting the graph using Matplotlib
plt.plot(x, y)
plt.show()


# Plotin the graph with labels and title
plt.plot(x, y)
plt.xlabel('Values')
plt.ylabel('sin-value')
plt.title('SIN Wave')
plt.show()
```

```python
plt.plot(x, z)
plt.show()


# Plotin the graph with labels and title
plt.plot(x, z)
plt.xlabel('Values')
plt.ylabel('cos-value')
plt.title('COS Wave')
plt.show()
```

```python
plt.plot(y, z)
plt.show()


# Plotin the graph with labels and title
plt.plot(y, z)
plt.xlabel('sin-values')
plt.ylabel('cos-values')
plt.title('SIN - COS GRAPHL')
plt.show()
```

```python
# Parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.title('PARABOLA GRAPHL')
plt.plot()
```

```python
plt.plot(x, y, 'r*')
plt.show()
```

```python
plt.plot(x, y, 'b_')
plt.show()
```

```python
x = np.linspace(-10, 10, 50)
plt.plot(x, np.sin(x), 'r-', label='sin')
plt.plot(x, np.cos(x), 'g--', label='cos')
plt.show()
```

```python
# BAR PLOT
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['ENGLISH', 'HINDI', 'MARATHI', 'TAMIL', 'TELUGU']
people = [75, 90, 60, 30, 45]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('% OF PEOPLES')
plt.show()
```

```python
fig = plt.Figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['ENGLISH', 'HINDI', 'MARATHI', 'TAMIL', 'TELUGU']
people = [75, 90, 60, 30, 45]
ax.pie(people, labels=languages)
plt.show()      
```

```python
# Scatter plot 
x = np.linspace(0,10, 50)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='r')
plt.show()
```

```python
# 3D Scatter plot
fig3 = plt.figure()
ax = fig3.add_axes([0, 0, 1, 1], projection='3d')
x = 20 * np.random.random(100)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x, y, z, c=x, cmap='Blues')
plt.show()
```
