```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Placement_Dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

### Central Tendencies
- Mean  -> Average of the dataset
- Median -> Middle value in sorted order and central tendency for odd number of values
- Mode   -> Most frequent value in the dataset


```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.distplot(df.salary)
```

```python
# Replce the missing values with Median value 
df['salary'].fillna(df['salary'].median(), inplace=True)
df.head()

# Output -> NaN replace with the median value 265000.0
```

```python
df2 = pd.read_csv('Placement_Dataset.csv')
df2.head()
```

```python
df2['salary'].fillna(df2['salary'].mean(), inplace=True)
df2.head()

# Output -> NaN value replaced by Mean value 288655.405405
```

```python
df3 = pd.read_csv('Placement_Dataset.csv')
df3['salary'].fillna(df3['salary'].mode()[0], inplace=True)
df3.head()

# Output -> NaN value replaced by Mode value 300000
```
