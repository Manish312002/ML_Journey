# SONAR ROCK VS MINE PREDICTION PROJECT

 predict whether an object is either Rock or Mine with SONAR Data. For this use case, we are using Logistic Regression Model for our prediction. 
 

## Set up for project


```python
# Import the Dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data collection and Data Processing


```python
# load the dataset
df = pd.read_csv('Sonar.csv', header=None)
    
```

```python
# Number of rows and columns
df.shape
```

```python
# Statistical Summary
df.describe()
```

```python
# Count the number of occurrences of each class
df[60].value_counts()
```

```python
# Group the features by class and compute the mean
df.groupby(60).mean()
```

```python
# Separating data and labels
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
print("X DATA SET :\n",X)
print("Y DATA SET :\n",Y)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print("X_Data : ",X.shape, X_train.shape, X_test.shape)
print("Y_Data : ",Y.shape, Y_train.shape, Y_test.shape)
```

```python
# Training the Logistic Regression Model
model = LogisticRegression()

model.fit(X_train, Y_train)
```

## Model Evaluation

```python
# Accuracy on the training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
print("Accuracy on Training data : ",training_data_accuracy)
```

```python
# Accuracy on the testing data
Y_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test_pred, Y_test)
print("Accuracy on testing data : ",testing_data_accuracy)
```
